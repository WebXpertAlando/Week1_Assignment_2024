```python
#Load the Necessary Libraries

import pandas as pd

import numpy as np
```

#### 1. Find all records where the weather was exactly clear?


```python
# Mainly Clear Weather
clear_weather = weather_data[weather_data['Weather'] == 'Mainly Clear']

clear_weather
```

#### 2. Find the number of times the wind speed was exactly 4 km/hr.

```python
wind_speed = weather_data[weather_data['Wind Speed_km/h'] =='4']

wind_speed.shape[0]
```

#### 3. Check if there are any NULL values present in the dataset

#### 4.	Rename the column "Weather" to "Weather_Condition."?

```python
# Renaming Column Weather to Weather_Condition

rename_column = weather_data.rename(columns={'Weather': 'Weather_Condition'})

rename_column
```

#### 5.	What is the mean visibility of the dataset?


```python
visibility_mean = weather_data['Visibility_km'].mean()

print ('The Visibility mean is:', visibility_mean)
```

#### 6.	Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
count_record = weather_data[(weather_data['Wind Speed_km/h'] > 24) & (weather_data['Visibility_km'] == '25')]

count_record2 = count_record.shape[0]

count_record2
```

#### 7.	What is the mean value of each column for each weather condition?

```python
mean_value = weather_data.mean()

mean_value
```

#### 8.	Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.


```python
weather_filtering = weather_data[(weather_data['Weather'] == 'Mainly Clear') & (weather_data['Rel Hum_%'] > 50) | (weather_data['Visibility_km'] > 40)]

weather_filtering
```
